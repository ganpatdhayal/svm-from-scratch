```python
import numpy as np
import matplotlib.pyplot as plt
from cvxopt import matrix, solvers

# Load dataset
def load_coffee_data():
    rng = np.random.default_rng(2)
    x = rng.random(400).reshape(-1, 2)
    x[:, 1] = x[:, 1] * 4 + 11.5
    x[:, 0] = x[:, 0] * (285 - 150) + 150
    Y = np.zeros(len(x))

    for i, (t, d) in enumerate(x):
        y = -3 / (260 - 175) * t + 21
        if 175 < t < 260 and 12 < d < 15 and d <= y:
            Y[i] = 1
        else:
            Y[i] = 0

    return x, Y.reshape(-1, 1)

# Kernel: RBF (Gaussian)
def rbf_kernel(X1, X2, sigma=5.0):
    X1_sq = np.sum(X1**2, axis=1).reshape(-1, 1)
    X2_sq = np.sum(X2**2, axis=1).reshape(1, -1)
    dist_sq = X1_sq + X2_sq - 2 * np.dot(X1, X2.T)
    return np.exp(-dist_sq / (2 * sigma**2))

# Prepare data
X, y = load_coffee_data()
y = y * 2 - 1  # Convert labels to {-1, 1}
n = X.shape[0]
K = rbf_kernel(X, X, sigma=10.0)  # You can try sigma = 1 to 10

# Set up parameters for QP
P = matrix(np.outer(y, y) * K)
q = matrix(-np.ones((n, 1)))
G = matrix(np.vstack((-np.eye(n), np.eye(n))))
h = matrix(np.vstack((np.zeros((n, 1)), np.ones((n, 1)) * 1.0)))  # C = 1.0
A = matrix(y.T.astype(np.double))
b = matrix(np.zeros(1))

# Solve QP
solvers.options['show_progress'] = False
sol = solvers.qp(P, q, G, h, A, b)
alphas = np.array(sol['x'])

# Get support vectors
sv = (alphas > 1e-5).flatten()
X_sv = X[sv]
y_sv = y[sv]
alpha_sv = alphas[sv]

# Compute bias using one support vector
def compute_bias():
    K_sv = rbf_kernel(X_sv, X_sv)
    return np.mean(y_sv - np.sum(alpha_sv * y_sv * K_sv, axis=1))

b = compute_bias()

# Predict
def project(X_test):
    K_test = rbf_kernel(X_test, X_sv, sigma=10.0)
    return np.dot(K_test, alpha_sv * y_sv) + b

def predict(X_test):
    return np.sign(project(X_test))

# Accuracy
y_pred = predict(X)
accuracy = np.mean(y_pred == y.flatten())
print("Training accuracy:", accuracy * 100, "%")

# Plot decision boundary
plt.figure(figsize=(8, 6))
plt.scatter(X[y.flatten() == -1][:, 0], X[y.flatten() == -1][:, 1], color='red', label='Class 0')
plt.scatter(X[y.flatten() == 1][:, 0], X[y.flatten() == 1][:, 1], color='blue', label='Class 1')
plt.scatter(X_sv[:, 0], X_sv[:, 1], s=100, facecolors='none', edgecolors='k', label='Support Vectors')

# Plot contour
xx, yy = np.meshgrid(np.linspace(X[:, 0].min(), X[:, 0].max(), 300),
                     np.linspace(X[:, 1].min(), X[:, 1].max(), 300))
grid = np.c_[xx.ravel(), yy.ravel()]
Z = predict(grid).reshape(xx.shape)
plt.contourf(xx, yy, Z, alpha=0.2, levels=[-1, 0, 1], colors=['red', 'blue'])
plt.contour(xx, yy, Z, colors='k', levels=[0], linewidths=2)

plt.title("SVM Decision Boundary with RBF Kernel (cvxopt)")
plt.xlabel("Temperature")
plt.ylabel("Roasting Time")
plt.legend()
plt.grid(True)
plt.show()
```
